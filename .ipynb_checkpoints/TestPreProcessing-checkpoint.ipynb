# Lower Casing English Words 

Lower casing is only apply for the English words. Having same word in upper case and lower case can reduce the accuracy. So as a first step convert all the english words to lower case.

```python
df_data2 = df_data.copy()
```

After converting upper case into lower case

```python
df_data['Statement_lower']= df_data['Statement'].apply(lambda x: x.lower())
df_data.head(20)
```

```python
df_data3 = df_data2.copy()
```

```python
from nltk.tokenize import sent_tokenize
```

After tokenizing sentences

```python
df_data2["Statement_tokened"] = df_data2["Statement"].apply(lambda x: sent_tokenize(x))
df_data2.head(20)
```

```python
df_data3["Statement"] = df_data3["Statement"].apply(lambda x: sent_tokenize(x))
```

Create seperate dataframe rows for the newly tokenized sentences.

# Removel Of Punctuation

Punctuations are removing, So words can be use to train the model. string.punctuation used to identify the punctuations.

```python
df_data4 = df_data3.copy()
```

After removing punctuations

```python
df_data3['Statement_punc_r']= df_data3['Statement'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df_data3.head(20)
```

```python
df_data4['Statement']= df_data4['Statement'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
```

# Replace Emojis

Emojis have their own meanings. Emojis can use to predic the negative and positive comments. So emojis are changed into respective words in this part. To identify emojis , demoji package is using here

```python
df_data5 = df_data4.copy()
```

```python
!pip install demoji
```

```python
import demoji
```

Identify emojies

```python
emo = df_data4['Statement'].apply(lambda x: demoji.findall(x))
emo.head(20)
```

```python
list_emos = list(emo)
list_filt = list(filter(None, list_emos))
list_filt
```

create seperate lists for emojies and their meanings

```python
list_emojis = []
list_emo_texts = []

for item in list_filt:
  list_emojis += list(item.keys())
  list_emo_texts += list(item.values())

print(list_emojis)
print(len(list_emojis))
print(list_emo_texts)
print(len(list_emo_texts))
```

Using above emoji and meaning lists, replace the emojies in the sentences with the relevant text.

```python
def replace_emoji(text):
  for i, item in enumerate(list_emojis):
    text = text.replace(item, " " + list_emo_texts[i] + " ")
  return text
```

Removing emojis and dataset after removing emojis

```python
df_data4['Statement_emo'] = df_data4['Statement'].apply(lambda x: replace_emoji(x))
df_data4.head(20)
```

# Removal of Frequent words

Most of the frequently used data are have none or littel semantic meaning related to a context. Most of them are articles, pronouns, or conjuctions. So removed top 10 most frequents words.

```python
df_data7 = df_data6.copy()
```

```python
from collections import Counter

cnt = Counter()

def idf_frq_wrds(text):
  for word in text.split():
    cnt[word] += 1

df_data6["Statement"].apply(lambda x: idf_frq_wrds(x))

cnt.most_common(10)
```

Removing frequent words and dataset after removing frequent words

```python
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])

def remove_f_words(text):
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df_data6["Statement_fwrds-rem"] = df_data6["Statement"].apply(lambda text: remove_f_words(text))
df_data6.head(20)
```

```python
df_data7["Statement"] = df_data7["Statement"].apply(lambda text: remove_f_words(text))
```

# Stemming

Process of removing inflected or sometimes derived words from the word stem. In this process sometime orignial word meaning is also lose.

```python
df_data8 = df_data7.copy()
```

Since this dataset have many sinhala words, I used sinhala langauge related package called singling.

https://github.com/rksk/sinhala-news-analysis/tree/master/sinhala-stemmer

```python
!pip install sinling
```

```python
from sinling import SinhalaStemmer
```

After stemming sinhala words, as we can see below there are lot of meaningless words creatted. So I did not take that dataframe for the further processes.

```python
stemmer = SinhalaStemmer()

def sin_stem_wrds(text):
  text_stem = ""
  for word in text.split():
    s_text =' '.join([str(x) for x in stemmer.stem(word)])
    text_stem = text_stem + " " + s_text
  return text_stem

df_data7["Statement_Stem"] = df_data7["Statement"].apply(lambda x: sin_stem_wrds(x))
df_data7.head(20)
```

Look like this is not a good stremmer. So did not apply changes for the further steps.

# Naive Bayes Model

Apply Naive Bayes model

Create x and y data from prepreocessed data..

```python
x_data = df_data8['Statement']
y_data = df_data8['Impact']
```

Split text and train data. 20% text data.

```python
pip install -U scikit-learn scipy matplotlib
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_data , y_data , stratify=y_data, test_size=0.2 )
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

vectorize x_train and x_test data.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer()
x_train = vec.fit_transform(x_train).toarray()
x_test = vec.transform(x_test).toarray()
```

Fit the data to model and train using x_train and y_train data

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train, y_train)
```

Evaluate model - Giving around 0.7 score

```python
model.score(x_test, y_test)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test , pred))
```
