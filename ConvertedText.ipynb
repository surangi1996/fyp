```python
import io
import os
import json
from google.cloud import vision
#from google.cloud.vision import types
```

```python
# Set up the Cloud Vision API client
client = vision.ImageAnnotatorClient()

# Set up the directory path and loop over the files
directory_path = 'E:/projects/image_scrape/downloads/tennis'
json_path = os.path.join(directory_path, 'json')
if not os.path.exists(json_path):
    os.makedirs(json_path)
for filename in os.listdir(directory_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):

        # Set up the image path and the JSON output path
        image_path = os.path.join(directory_path, filename)
        output_path = os.path.join(json_path, filename[:-4] + '.json')

        # Check if the JSON file already exists
        if not os.path.exists(output_path):

            # Read the image file and set up the image request
            with io.open(image_path, 'rb') as image_file:
                content = image_file.read()
            image = vision.Image(content=content)
            features = [vision.Feature(type=vision.Feature.Type.TEXT_DETECTION)]
            request = vision.AnnotateImageRequest(image=image, features=features)

            # Submit the image request and save the JSON response to a file
            response = client.annotate_image(request)
            with open(output_path, 'w') as output_file:
                output_file.write(vision.AnnotateImageResponse.to_json(response))
                #json.dump(response, output_file)
        
                  
                
            print(filename)
            
            if response.error.message:
                raise Exception(
                    '{}\nFor more info on error messages, check: '
                    'https://cloud.google.com/apis/design/errors'.format(
                        response.error.message))
        else:
            print("skip: " + filename)
```

# create CSV file

```python
import os
import json
import csv
```

```python
# Set up the parent directory path and the output CSV file
parent_path = 'E:\projects\image_scrape\downloads'
output_path = os.path.join(parent_path, 'labeled.csv')

# Loop through the subdirectories
with open(output_path, 'w', encoding="utf-8", newline='') as output_file:
    writer = csv.writer(output_file)
    writer.writerow(['index', 'file', 'text', 'sport'])
    index = 1
    for directory in os.listdir(parent_path):
        directory_path = os.path.join(parent_path, directory)

        # Check if the subdirectory has a 'json' folder
        json_path = os.path.join(directory_path, 'json')
        if not os.path.exists(json_path):
            print(f"No 'json' folder found in '{directory}' subdirectory.")
            continue

        # Loop through the JSON files in the 'json' folder
        for filename in os.listdir(json_path):
            if not filename.endswith('.json'):
                continue

            #print(json_path + '\\' + filename)
            
            # Read the JSON file and extract the text annotations
            json_pathname = os.path.join(json_path, filename)
            with open(json_pathname, 'r', encoding="utf-8") as json_file:
                json_data = json.load(json_file)
                #print(len(json_data['textAnnotations']))
                if len(json_data['textAnnotations']) > 0 :
                    text_annotations = json_data['textAnnotations'][0]['description'].replace('\n', ' ')
                else :
                    continue
                

            # Write the output to the CSV file
            writer.writerow([index, filename.replace('.json', ''), text_annotations, directory])
            index += 1
```
